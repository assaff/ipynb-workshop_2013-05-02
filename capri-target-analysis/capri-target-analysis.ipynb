### Per-target functions:

Given a target's score tables and a specified column name (`col_name`), this function selects the best (i.e. lowest energy in column `col_name`) decoy for each of the target's windows, and plots them against each other in a line plot.

It saves the plot as an image in that target's directory.

```python
def plot_by_column(target_name, target_score_tables, col_name):
    score_column = [sc.ix[:, col_name].order().head(1) for sc in target_score_tables]
    top_decoys = pd.concat(score_column)
    top_decoys.index = [decoy_name.split('.')[0] for decoy_name in top_decoys.index]
    top_decoys.plot(kind='line', lw=2, marker='o', markersize=4)
    pl.ylim(top_decoys.mean()-2* top_decoys.std(), top_decoys.mean()+2*top_decoys.std(),)
    pl.ylabel(col_name)
    pl.savefig('{target_name}/{target_name}_sliding-window_{col}.svg'.format(target_name=target_name, col=col_name))
    return top_decoys
```

This function runs on a *target* context.

It basically:

1. reads all the sliding-windows for that target
2. loads the score table for each of them
3. plots the result for the best decoy from each window, over each of the columns in the `important_cols` list.

**The result should be one plot comparing the windows of that target for each of the `important_cols`**.

```python
def plot_target_decoys(target_name):
    wins_dataframe = pd.read_csv(
                        '{target_name}/seqs.{target_name}'.format(target_name=target_name), 
                        header=None,
                    )
    wins = wins_dataframe[0].values
    score_tables = [pd.read_csv('{target_name}/{win_name}/score.sc'.format(target_name=target_name, win_name=win), 
                            delim_whitespace=True, 
                            index_col='description', 
                            skiprows=1) for win in wins]
    
    print 'Target: {target_name}, Subsequences: {sequences}'.format(target_name=target_name,
                                                                    sequences=' '.join(wins))
    
    for i, col_name in enumerate(important_cols, 1):
        #pl.subplot(3,1,i)
        plot_by_column(target_name, score_tables, col_name, )
        pl.show()
    
```
