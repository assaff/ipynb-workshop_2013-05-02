The table is stored locally as a `.csv` file, which stands for *comma-separated values*.

```python
athletes = pd.read_csv(
            'All London 2012 athletes - ALL ATHLETES.csv', 
            parse_dates=['Date of birth'], 
            sep=',',
            )
```

Let's see a brief description of the `athletes` object:

```python
athletes # alternatively, we could run display(athletes)
```

`athletes` is initialized as a *DataFrame* object, which is essentially a table with some very nifty features. It is the most-often used data structure in Pandas.

You can also see that some columns (e.g. `Place of birth`) have some null values. Pandas is very good at handling this kind of irregularity.

We can also see which data types were assigned to each of the columns using the `dtypes` attribute/field. Note that **`pandas` does that automatically**, so you don't have to.

```python
n_rows, n_cols = athletes.shape
print 'The table has %d rows and %d columns' % (n_rows, n_cols)

print
print athletes.dtypes
```

### Indexing/slicing using the `ix` method:

We're dealing with a very large table here, as we often do. The easiest way to narrow down the scope is to select a subset of the rows and/or columns.

You can do that using the `ix` method.

```python
top6by6 = athletes.ix[:6,:4]
top6by6
```

```python
gc = athletes.groupby('Country')
size_series = gc.size().tail(20)#.order(ascending=False).head(10)
size_series
```

```python
size_series.plot(kind='barh',) 
pl.xlabel('Number of athletes')
pl.show()
```

### Mean age by country

```python
group_by_country = athletes.groupby(['Country'])
mean_ages = group_by_country['Age'].mean()
mean_ages.order(ascending=False).head(20)#
```

### Medal distribution among nations

```python
medal_table = athletes.groupby(['Country'])[['Gold', 'Silver', 'Bronze']].sum().sort('Gold', ascending=False)
medal_table.head(5)
```

```python
total_series = medal_table.sum(axis=1).order(ascending=False)
print total_series.head(5)
```

```python
total_medals=athletes.groupby(['Country'])['Total'].sum()
```

```python
pl.hist(total_medals, bins=20, histtype='step', lw=3, color='green')

pl.yscale('symlog')

pl.xlabel('Number of medals won')
pl.ylabel('Number of countries')
pl.title('Distribution of medals among countries')

pl.savefig('histo_medals.png')

pl.show()
```

### Weight distribution among the sexes

```python
for sex_label, sex_group in athletes.groupby(['Sex'])['Weight']:
    pl.hist(sex_group, bins=range(20, 160, 5), label=sex_label, alpha=.7,  )
pl.legend()
pl.xlabel('Weight [Kg]')
pl.ylabel('count')
pl.title('Weight distributions among male and female athletes')
pl.show()
```
